```python
import pandas as pd
import numpy as np
import yfinance as yf
import matplotlib.pyplot as plt
```

https://finance.yahoo.com/

```python
data = yf.download(["BAC","BK","BBWI","BAX","BDX"], start="2017-01-01", end="2022-09-13")
data.tail()
```

```python
data = data[['Adj Close']]
```

```python
data.head()
```

```python
data.columns = ['BAC', 'BAX', 'BBWI','BDX','BK']
```

```python
returns = data.pct_change().dropna()
returns.head()
```

### Holding period Return

```python
data.plot()
plt.xlabel('Date')
plt.ylabel('Stock Price')
```

If we were to plot to share prices togtheer we would need to have two axes as they are at different levels and the above doesnt really show the variability in the lower price margin

The graph below shows the variabliity of the graph with respect to the stock price and date, for different companies.

```python
fig, ax1 = plt.subplots()
fig.set_size_inches(25,8)

color = 'tab:red'
ax1.set_xlabel('Date')
ax1.set_ylabel('Index',color = 'black')
ax1.plot(data['BAC'], color=color)


ax2 = ax1.twinx()  

color = 'tab:blue'
ax2.set_ylabel('$', color=color) 
ax2.plot(data['BAX'], color=color)


ax3 = ax1.twinx() 

color = 'tab:green' 
ax3.plot(data['BBWI'], color=color)


ax4  = ax1.twinx()
color = 'tab:purple'
ax4.plot(data['BDX'],color=color)

ax5  = ax1.twinx()
color = 'tab:orange'
ax5.plot(data['BK'],color=color)


fig.tight_layout()  
plt.show()
```

```python
#CAlculate the holding period return
# .iloc[-1] is the last obseration in the data set
# .iloc[0] is the first obseration in the data set
HPR_return = ((data.iloc[-1] - data.iloc[0]) / data.iloc[0])
HPR_return
```

HRP- Holding Period return for all the companies from 2017 to 2022 are BAC:76%, BAX:44%, BBWI: -7.3%, BDX:77%, BK:6.5%
The returns from BBWI has decreased in the past 5 years

```python
## now what is the average annual return
days = len(data.index)
annualized_return=((1 + HPR_return)**(252/days))-1
print (annualized_return)
```

The annual rise is the highest in BAC:10.4% and BDX:10.5%. BBWI has seen an annual decline of -1.3%

```python
# Your computed annualized returns
annualized_return = {'BAC': 0.104908, 'BAX': 0.066345,'BBWI': -0.013296, 'BDX': 0.105771, 'BK': 0.011162}

# Create a bar chart
plt.bar(annualized_return.keys(), annualized_return.values(), color=['blue', 'green','red','orange','purple'])

# Add labels and title
plt.xlabel('Assets')
plt.ylabel('Annualized Return')
plt.title('Bar Chart of Annualized Return for BAC, BAX, BBWI, BDX and BK')

# Show the plot
plt.show()
```

```python
# Calculate the daily returns
returns = data.pct_change()
```

```python
# Plot the returns
returns.plot(figsize=(14, 4))
plt.title('Daily Returns for Stock and Index')
plt.ylabel('Daily Return')
plt.xlabel('Date')
plt.grid(True)
plt.legend()
plt.show()
```

Its important to remember that the Standard deviation of price doesnt have menaing for us in the way that the standard deviation of returns does - as a measure of the riskiness of the stock.

```python
print("mean : ", returns.mean())
print("vol : ", returns.std())
print("skew : ", returns.skew())
print("kurt : ", returns.kurtosis())
```
